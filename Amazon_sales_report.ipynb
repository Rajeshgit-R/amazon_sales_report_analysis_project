```python
df.columns
```

```python
ax=sns.countplot(x="Size",data=df)
```

```python
ax=sns.countplot(x="Size",data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

**Note:Above graph you can see that most of the pepole buy M-size**

```python
#group by
df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
```

```python
size_Qty=df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
sns.barplot(x="Size",y="Qty",data=size_Qty)
```

**Note:Above the graph you can see that most of the Qty buy M-size in the sales**

```python
#courier status
sns.countplot(data=df,x="Courier Status",hue="Status")
```

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df,x="Courier Status",hue="Status")
```

**Note:From above the graph majority of the orders shipped through the courier**

```python
#histgram
df["Size"].hist()
```

```python
df["Category"]=df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor="black")
plt.xticks(rotation=90)
plt.show()
```

**Note:Above the graph you can see that most of the buyers are T-shirt**

```python
#check B2B data by using piechart
B2B_check=df["B2B"].value_counts()

#plot the piechart
plt.pie(B2B_check,labels=B2B_check,autopct="%1.1f%%")
#plt.axis("equal")
plt.show()
```

```python
# Checking B2B Data  by using pie chart
B2B_Check = df["B2B"].value_counts()

#  Plot the pie chart
plt.pie(B2B_Check, labels=B2B_Check.index, autopct="%1.1f%%")
#plt.axis("equal")
plt.show()
```

**Note:Abouve the piechart we can maximum i.e.99.2% of buyers are retailers and 0.8% B2B buyers**

```python
a1=df["Fulfilment"].value_counts()
fig, ax=plt.subplots()
plt.pie(a1,labels=a1.index,autopct="%1.1f%%",radius=0.7,wedgeprops=dict(width=0.6))
plt.show()
```

```python
#prepare data scatter plot
x_data=df["Category"]
y_data=df["Size"]

plt.scatter(x_data,y_data)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("Scatter plot")
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="ship-state")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("Distribution of state")
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 state
top_10_state=df["ship-state"].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.countplot(data=df[df["ship-state"].isin(top_10_state.index)],x="ship-state")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("Distribution of state")
plt.xticks(rotation=45)
plt.show()
```

**Note:Above the graph you can see that most of the buyers are Maharashtra state**
